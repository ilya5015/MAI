```python
import pygmo as pg
import numpy as np
import pandas as pd
```

```python
class sphere_function:
    def fitness(self, x):
        return [sum(x*x)]

    def get_bounds(self):
        return ([-1,-1],[1,1])
```

```python
class matyas_function:
    def fitness(self, x):
        return [0.26 * (sum([xi**2 for xi in x])) - np.prod([np.cos(xi) for xi in x])]

    def get_bounds(self):
        return ([-10.0] * 10, [10.0] * 10)
```

```python
# Настройка алгоритмов
def optimize(func, algo, **kwargs):
    # Создаем проблему
    prob = pg.problem(func)
    # Создаем популяцию
    pop = pg.population(prob, size=20)
    # Запускаем эволюцию
    pop = algo.evolve(pop)
    return pop.get_f()[0][0]  # Возвращаем значение фитнеса
```

```python
# Оптимизация функции сферы
sphere_algorithms = {
    'Differential Evolution': pg.algorithm(pg.de()),
    'Particle Swarm Optimization': pg.algorithm(pg.pso()),
    'CMA-ES': pg.algorithm(pg.cmaes())
}

results_sphere = {}
for name, algo in sphere_algorithms.items():
    result = optimize(sphere_function(), algo)
    results_sphere[name] = result
```

```python
# Оптимизация функции Матьяса
sphere_algorithms = {
    'Differential Evolution': pg.algorithm(pg.de()),
    'Particle Swarm Optimization': pg.algorithm(pg.pso()),
    'CMA-ES': pg.algorithm(pg.cmaes())
}

results_matyas = {}
for name, algo in sphere_algorithms.items():
    result = optimize(matyas_function(), algo)
    results_matyas[name] = result
```

```python
# Создание таблицы сравнения
comparison_table = pd.DataFrame({
    'Algorithm': list(results_sphere.keys()),
    'Sphere Function': list(results_sphere.values()),
    'Matyas Function': list(results_matyas.values())
})
```
